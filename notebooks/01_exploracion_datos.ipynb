# Exploración de Datos: Trigo y Clima

Notebook para exploración inicial del dataset maestro.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

ROOT = Path.cwd().parent
print(f"ROOT: {ROOT}")
```

## Cargar dataset

```python
df = pd.read_csv(ROOT / "data" / "processed" / "dataset_maestro_ia.csv")
print(f"Shape: {df.shape}")
df.head()
```

## Estadísticas descriptivas

```python
df.describe()
```

## Distribución del rendimiento

```python
plt.figure(figsize=(10, 5))
plt.hist(df['RENDIMIENTO - KG X HA'].dropna(), bins=50, edgecolor='black')
plt.xlabel('Rendimiento (kg/ha)')
plt.ylabel('Frecuencia')
plt.title('Distribución del rendimiento de trigo')
plt.show()
```

## Correlaciones entre variables

```python
# Seleccionar solo columnas numéricas de clima
clima_cols = [c for c in df.columns if c.startswith(('lluvia_', 'tmax_', 'tmin_'))]
corr = df[clima_cols + ['RENDIMIENTO - KG X HA']].corr()['RENDIMIENTO - KG X HA'].sort_values(ascending=False)
print("Top 10 correlaciones con rendimiento:")
print(corr.head(11))  # 11 porque incluye la correlación consigo mismo
```
